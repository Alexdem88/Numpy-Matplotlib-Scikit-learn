```python
boston = load_boston()
```

```python
for line in boston.DESCR.split('\n'):
    print(line)
```

```python
data=boston.data
```

```python
feature_names=boston.feature_names
```

```python
X= pd.DataFrame(data, columns = feature_names)
```

```python
X.head()
```

```python
X.info()
```

```python
price=boston.target
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,price, test_size=0.2, random_state=42)
```
